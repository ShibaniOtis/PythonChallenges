````python
num = int(input("Choose a number to check whether it is prime or not"))

def check_prime(num):

    divisors = []

    for i in range(1, num+1):
        if num % i == 0:
            divisors.append(i)

    if len(divisors) == 2:     # A prime number is only divisble by 1 and itself, hence 2 factors
        print(str(num) + " is a prime number.")
    else:
        print(str(num) + " is not a prime number.") 

check_prime(num)
````

# Fibonacci

## Exercise 13

Write a program that asks the user how many Fibonacci numbers to generate and then generates them. Take this opportunity to think about how you can use functions. Make sure to ask the user to enter the number of numbers in the sequence to generate.(*Hint: The Fibonacci seqence is a sequence of numbers where the next number in the sequence is the sum of the previous two numbers in the sequence. The sequence looks like this: 1, 1, 2, 3, 5, 8, 13, …*)

<br/>

## Solution:

````python
num = int(input("How many Fibonacci numbers would you like to generate?"))

def fibonacci(num):

    seq = []

    for i in range(num):
        if i == 0 or i == 1:  # The first two fibonacci numbers are 1, hence the two cases
            seq.append(1)
        else:
            # Note: seq of numbers where the next number in the seq = the sum of the previous 2 numbers
            seq.append(seq[i-1] + seq[i-2])    # for i >= 2 case
    return seq

fibonacci(num)
````

# Password Generator

## Exercise 16

Write a password generator in Python. Be creative with how you generate passwords - strong passwords have a mix of lowercase letters, uppercase letters, numbers, and symbols. The passwords should be random, generating a new password every time the user asks for a new password. Include your run-time code in a main method.

Extra:

- Ask the user how strong they want their password to be. For weak passwords, pick a word or two from a list.

<br>

## Solution:

````python
import random
import string

strength = str(input("Do you want your password to be strong or weak?")).lower()

def password_gen(strength):

    # produces a list of 5 random numbers of a random length b/w 3 and 5
    rand_num = [str(random.randint(0,9)) for i in range(random.randint(3,5))]

    # produces a list of 5 random letters, regardless of case of a random length b/w 3 and 5
    rand_letter = [random.choice(string.ascii_letters) for i in range(random.randint(3,5))]

    # produce a list of 5 random symbols of a random length b/w 3 and 5
    rand_symbol = [random.choice(string.punctuation) for i in range(random.randint(3,5))]

    # concatenate everything, shuffle the 3 lists 
    mix = rand_num + rand_letter + rand_symbol
    random.shuffle(mix)

    wordlist = ["Cat", "Dog", "Bird", "Sunshine", "Password", "Love", "Cuckoo", "Stinky"]
    
    if strength == "strong" or strength == "s":
        return "".join(mix)         # Join each shuffled element to form a random password
    elif strength == "weak" or strength == "w":
        return random.choice(wordlist) + " " + random.choice(wordlist)  # Chooses 2 random words from the wordlist
    else:
        return "Invalid input! Try again."

print(password_gen(strength))
````

Randomly generate a 4-digit number. Ask the user to guess a 4-digit number. For every digit that the user guessed correctly in the correct place, they have a “cow”. For every digit the user guessed correctly in the wrong place is a “bull.” Every time the user makes a guess, tell them how many “cows” and “bulls” they have. Once the user guesses the correct number, the game is over. Keep track of the number of guesses the user makes throughout the game and tell the user at the end. Use a "main" method in your program.

Say the number generated by the computer is 1038. An example interaction could look like this:

 > ``` 
 Welcome to the Cows and Bulls Game! 
 Enter a number:
 >>>> 1234                           
 2 cows, 0 bulls
 >>>> 1286                           
 1 cow, 1 bull                       
 ...
 > ```

Until the user guesses the number.

<br>

## Solution:

````python
import random

def cows_and_bulls():

    # generates random 4-digit number
    randfour = random.randint(1000, 9999)
    randfour = str(randfour)  # need to convert the number to a string for the for loop 
    
    guesses = 0

    print("Welcome to the Cows and Bulls Game!")

    while True:

        cows = 0
        bulls = 0
        guess = str(input("Enter a 4-digit number:")) 

        print(">>>", guess)

        if len(guess) != 4:
            return "Uh-oh! You needed to enter a 4-digit number... Try again."  

        for i in range(0, len(randfour)): 
            if randfour[i] == guess[i]:
                cows += 1
            elif guess[i] in randfour:
                bulls += 1
        guesses += 1

        print(str(cows) + " cows, " + str(bulls) + " bulls \n" + "Number of guesses: " + str(guesses))

        if guess == randfour:
            return "Congratulations you win!"

# Main method like C++ uses (more useful if we wanted to save this as a .py file)
def main():

    print(cows_and_bulls())

if __name__ == "__main__":
    main()
````

## Exercise 20

Write a function that takes an ordered list of numbers (a list where the elements are in order from smallest to largest) and another number. The function decides whether or not the given number is inside the list and returns (then prints) an appropriate boolean.

<br>

## Solution:

````python
# Program WITHOUT doing a "binary search":

import random

# Generate a list of 100 random elements/numbers
randls = sorted([random.randint(1,1000) for i in range(100)])

def element_search():
    while True:
        num = int(input("Enter a number to check if it is in the list:"))

        if num in randls:
            return str(num) + " is in the list."
        else:
            return str(num) + " is not in the list"
    
element_search()


# Program using a binary search:

# Let's say we have an element (number) and an ordered list. A binary search is knowing how the element you are looking for compares to another element in the list, splits the list in two - one half where the element can be, and one where it definitely cannot be. In the case where our list contains millions of elements, knowing that we can cut the “search space” in half is a great increase in efficiency.

randls = sorted([random.randint(1,1000) for i in range(100)])

def element_binarysearch():

    # Find the index of the middle element of the list
    middle = int(len(randls)/2)

    # Make a copy of randls
    newls = randls[:]

    found = False

    num = int(input("Enter a number to check if it is in the list:"))

    while len(newls) > 1:
        
        if num < newls[middle]:
            newls = newls[:middle]   # newls is everything to the left of the middle value 
        elif num > newls[middle]:
            newls = newls[middle:]   # newls is everything to the right of the middle value
        else:
            found = True
            break
        print(newls)
        middle = int(len(newls)/2)  # new middle index
        
    if found:
        return str(num) + " is in the list!"
    else:
        return str(num) + " is not in the list!"

element_binarysearch()
````

Given a ```.txt``` file that has a list of a bunch of names, count how many of each name there are in the file, and print out the results to the screen. I have a ```nameslist.txt``` file for you, if you want to use it!

Extra:

- Take the ```training.txt``` file and count how many of each “category” of each image there are. This text file is actually a list of files corresponding to the SUN database scene recognition database, and lists the file directory hierarchy for the images. Once you take a look at the first line or two of the file, it will be clear which part represents the scene category. To do this, you’re going to have to remember a bit about string parsing in Python 3. I talked a little bit about it in this post.

<br>

## Solution:

````python
# Open the text file to read
open_file = open("nameslist.txt", "r")

# Read the file and seperate each name by new line
names = open_file.read().split("\n")
open_file.close()   # closing the file after reading it is good practice

# The get() method is specifically for dictionaries & returns a value for the given key. If key is unavailable then it returns       the default value "None".

# Create empty dictionary 
names_dict = {}

# Loop through each name in the list of names and add 1 to the value of each unique name (acts like a key). 
# Store the keys and values into our dictionary
for name in names:
    names_dict[name] = names_dict.get(name, 0) + 1  # dict.get(key,value). Specify a default value if none found (we chose 0)

print(names_dict)
   
````

## Extra Solution:

````python
# Open and read the training.txt file

open_file = open("training.txt", "r")

training = open_file.read().split("\n")  # split the text by new line into a list
open_file.close()
#print(training)

categories = {}

for x in training:
    
    ls = x.split("/")  # split each string in the list by "/" into substrings
    #print(ls)

    # Make sure the list from the split has at least as many elements to hold the category name
    # In our case, we had an empty string at the end of "training" and didn't have a 3rd element to get after the split, so we           need everything with 3 or more elements
    if len(ls) > 2:          
        cat = ls[2]   # want to grab the 3rd string in this list (category name)
        categories[cat] = categories.get(cat, 0) + 1   # use the category names (key) to make the dictionary
    
print(categories)
````

Given two ```.txt``` files that have lists of numbers in them, find the numbers that are overlapping. The ```primenumbers.txt``` file has a list of all prime numbers under 1000, and the ```happynumbers.txt``` file has a list of happy numbers up to 1000.

(If you forgot, prime numbers are numbers that can’t be divided by any other number. And yes, happy numbers are a real thing in mathematics - you can look it up on Wikipedia.
<br>

## Solution:

````python
# Let's open two files
with open("primenumbers.txt", "r") as f1, open("happynumbers.txt", "r") as f2:
    primenums = f1.read().split("\n")
    happynums = f2.read().split("\n")
    
f1.close()
f2.close()

overlapls = []  # empty list to store our happy primes

for p in primenums:
    for h in happynums:
        if p == h:
            overlapls.append(p)

print(overlapls)

# List comprehension for this: 
overlapls = [elem for elem in primenums if elem in happynums]
print(overlapls)
````

# Guessing Game Two 

## Exercise 25

In a previous exercise, we’ve written a program that “knows” a number and asks a user to guess it.

This time, we’re going to do exactly the opposite. You, the user, will have in your head a number between 0 and 100. The program will guess a number, and you, the user, will say whether it is too high, too low, or your number.

At the end of this exchange, your program should print out how many guesses it took to get your number.

As the writer of this program, you will have to choose how your program will strategically guess. A naive strategy can be to simply start the guessing at 1, and keep going (2, 3, 4, etc.) until you hit the number. But that’s not an optimal guessing strategy. An alternate strategy might be to guess 50 (right in the middle of the range), and then increase / decrease by 1 as needed. After you’ve written the program, try to find the optimal strategy (binary search)! 

This exercise doesn’t need any functions, but it does need extensive use of variables, math, and user input.

<br>

## Solution:

````python
# The best strategy for the computer to guess the number in your head is by using a "binary search algorithm":

low = 0
high = 100

guesses = 0  

while True:

    # We need to add "low" to the difference cut in half, so that the middle number is between the new low and high  
    middle = int((high - low) / 2) + low    

    # Program starts by guessing the middle number
    print("Is your number " + str(middle) + "?")   

    guesses+=1

    reply = input("Too high, too low, or correct:").lower().strip()  # get's rid of white space before or after the response typed

    if reply == "too high":
        high = middle - 1   # Since the middle is too high of a guess, we need the new high to be the number before the middle

    elif reply == "too low":
        low = middle + 1    # Since the middle is too low of a guess, we start guessing from the number after the middle (new low)

    elif reply == "correct":
        print(":-)")
        break

print("Number of guesses: " + str(guesses))
````

# Check Tic Tac Toe 

## Exercise 26

*\*This exercise is Part 2 of 4 of the Tic Tac Toe exercise series.*

As you may have guessed, we are trying to build up to a full tic-tac-toe board. However, this is significantly more than half an hour of coding, so we’re doing it in pieces.

Today, we will simply focus on checking whether someone has WON a game of Tic Tac Toe, not worrying about how the moves were made.

If a game of Tic Tac Toe is represented as a list of lists, like so:

```
game = [[1, 2, 0],
	[2, 1, 0],
	[2, 1, 1]]
```
where a 0 means an empty square, a 1 means that player 1 put their token in that space, and a 2 means that player 2 put their token in that space.

Your task this week: Given a 3 by 3 list of lists that represents a Tic Tac Toe game board, tell me whether anyone has won, and tell me which player won, if any. A Tic Tac Toe win is 3 in a row - either in a row, a column, or a diagonal. Assume that in every board there will only be one winner.

Here are some more examples to work with:
```
winner_is_2 = [[2, 2, 0],
	       [2, 1, 0],
	       [2, 1, 1]]

winner_is_1 = [[1, 2, 0],
	       [2, 1, 0],
	       [2, 1, 1]]

winner_is_also_1 = [[0, 1, 0],
		    [2, 1, 0],
		    [2, 1, 1]]

no_winner = [[1, 2, 0],
	     [2, 1, 0],
	     [2, 1, 2]]

also_no_winner = [[1, 2, 0],
	    	  [2, 1, 0],
	    	  [2, 1, 0]]
```
<br> 

## Solution:

````python
# Let's use a given list of lists (matrix) to test on  
game = [[1, 2, 0],
	    [2, 0, 1],
	    [0, 2, 1]]

# Let's check for all possible winning moves:
def check_for_winner(game):

    for i in range(len(game)):   # range is 0 to 3 since game board is 3 by 3
        if game[i][0] == game[i][1] == game[i][2] != 0:    # across the rows
            return game[i][0]
                                              
        elif game[0][i] == game[1][i] == game[2][i] != 0:  # down the columns
            return game[0][i]

    if game[0][0] == game[1][1] == game[2][2] != 0:        # along diagonal
        return game[0][0]
        
    elif game[0][2] == game[1][1] == game[2][0] != 0:      # along other diagonal
        return game[0][2]

winner = check_for_winner(game)

if winner:
    print("Player " + str(winner) + " won!")
else:
    print("No one won.")
````

As a reminder, our tic tac toe game is really a list of lists. The game starts out with an empty game board like this:

And ask Player 2 for their move, printing an ```O``` in that place.

- Notice how in the example I gave coordinates for where I want to move starting from (1, 1) instead of (0, 0). <br>

## Solution:

````python
def draw_board(game):

    size = len(game)  # The game has 3 lists, so board size will be 3 by 3

    moves = {0: " ", 1: "X", 2: "O"} # Create a dictionary of possible moves

    # We will create the board layout as we go:
    for g in game:       
        print(" ---" * size)    # hard code the top of the game board and between each row

        moves_ls = []    # create a list to store the moves as we come across them in the game

        for elem in g:   # elements in each sublist
            moves_ls.append(moves[elem])
        print("| " + " | ".join(moves_ls) + " |") 

        # Using list comprehension: print("| " + " | ".join([moves[elem] for elem in g]) + " |")    

    print(" ---" * size)        # hard code the bottom of the game board

    ### Alternative but longer way: 

    # for g in game:       
    #    print(" ---" * size)
        # s = "| "
        # for i in g:
        #     if i == 0:
        #         s+= " "
        #     elif i == 1: # player 1 is X
        #         s+= "X"
        #     elif i == 2: # player 2 is O
        #         s+= "O"  
        #     s+= " | "    # the seperation between each box
        # print(s)
    # print(" ---" * size)
        
# Test case, comes before the function is called:
game = [[1, 1, 2],
        [2, 1, 2],
        [0, 2, 1]]

#draw_board(game)

def update_board(game, player):

    while True:

        reply = input("Let's play Tic Tac Toe!  What is your move? (Type your move as \"row,col\")").strip()

        coords = reply.split(",") # list with chosen row and column as strings 
        row = int(coords[0]) - 1
        col = int(coords[1]) - 1

        if game[row][col] == 0:   # valid move i.e. no X or O in that row and col of the game board  
            break                 # the while loop stops and game allows the player to place X or O

        print("Invalid move. Try again!")  # if game[row][col] != 0 i.e. there is already X or O in that spot, loop continues

    game[row][col] = player       # where the player places his/her move


def check_game_over(game):

    zeroes = 0

    for g in game:
        zeroes += g.count(0)  # count all the 0's on each row of the game

    if zeroes == 0:         # the game board is full (i.e no zeroes)
        print("Game over!")
        return True
    else:                   # the game board has empty spots
        return False


# Test case:
game = [[0, 0, 0],
        [0, 0, 0],
        [0, 0, 0]]

draw_board(game)  # print empty game board

while True:
    
    update_board(game, 1) # call player 1
    draw_board(game)  # print updated game board

    # NOTE: the game ends with the first player (makes last move)
    if check_game_over(game):     
        break

    update_board(game, 2) # call player 2
    draw_board(game)  
````

An example interaction can look like this:

```
>>> Welcome to Hangman!
_ _ _ _ _ _ _ _ _

>>> Guess your letter: S 
Incorrect!
>>> Guess your letter: E

E _ _ _ _ _ _ _ E
...

```

And so on, until the player gets the word.

<br>

## Solution:

````python
# Let's first read the SOWPODS text file
open_file = open("sowpods.txt", "r")  
s = open_file.read()
open_file.close()

sowpods = s.split("\n")

import random
    
randWord = random.choice(sowpods)

print(">>> Welcome to Hangman!")

stage = ["_"] * len(randWord)  # List of "_". The stage/state will change with each correct guess

print(randWord)

# Sets look up an item quickly w/o checking all the other elements like a list would
guesses = set()   # keeps track of all the letters guessed (*sets store unique items)

while "_" in stage:

    guess = input("Guess your letter:").upper()

    if guess in guesses:
        print("You've already guessed that letter. Guess again!")
        continue     # skips running the rest of the while loop and goes back to the start 

    guesses.add(guess)  # With sets you use .add() and not .append()

    for i in range(len(randWord)):
        if guess == randWord[i]:    # if guess matches a letter in the random word
            stage[i] = randWord[i]  # replace the "_" in position "i" of the stage with the letter

    if guess not in randWord:
        print("Incorrect!")

    print(" ".join(stage) + "\n")  # We want to join the list of underscores and/or letters with spaces between them
````

````python
# This function will draw the hangman as you play!
def hangman(numguess):

    head = "O"
    arm1 = "/"
    body = "|"
    arm2 = "\\"
    leg1 = "/"
    leg2 = "\\"

    print("    __________")
    print("    |        |")
    if numguess >= 1:
        print("    |        " + head)
    else: 
        print("    |")

    if numguess >= 4:
        print("    |       " + arm1 + body + arm2)
    elif numguess >= 3:
        print("    |       " + arm1 + body)
    elif numguess >= 2:
        print("    |       " + arm1) 
    else:
        print("    |")  

    if numguess >= 6:
        print("    |       " + leg1 + " " + leg2 + "   HANGMAN!")
    elif numguess >= 5:
        print("    |       " + leg1)
    else:
        print("    |")
    print("    |")
    print("    |")
    print("    |")
    print(" ___|___\n")

# Let's first read the SOWPODS text file
open_file = open("sowpods.txt", "r")  
s = open_file.read()
open_file.close()

sowpods = s.split("\n")

import random

while True:   # restarts the game

    randWord = random.choice(sowpods)

    print(">>> Welcome to Hangman!")

    stage = ["_"] * len(randWord)  # List of "_". The stage/state will change with each correct guess

    print(randWord)

    # Sets look up an item quickly w/o checking all the other elements like a list would
    guesses = set()   # keeps track of all the letters guessed (*sets store unique items)

    numguess = 0

    while "_" in stage and numguess < 6:

        guess = input("Guess your letter:").upper()
    
        if guess in guesses:
            print("You've already guessed that letter. Guess again!")
            continue     # skips running the rest of the while loop and goes back to the start 

        guesses.add(guess)  # with sets you use .add() and not .append()

        for i in range(len(randWord)):
            if guess == randWord[i]:    # if guess matches a letter in the random word
                stage[i] = randWord[i]  # replace the "_" in position "i" of the stage with the letter

        if guess not in randWord:
            numguess += 1
        
        hangman(numguess)   # we call the function at the end and use num of guesses so far to create the hangman

        print(" ".join(stage) + "\n")  # we want to join the list of underscores and/or letters with spaces between them

    again = input("Do you want to play again? y/n?").lower()

    if again == "no" or again == "n":
        print("Goodbye!")
        break
````

# Birthday Plots

## Exercise 36

*This exercise is Part 4 of 4 of the birthday data exercise series.*

In the previous exercise we counted how many birthdays there are in each month in our dictionary of birthdays.

In this exercise, use the ```bokeh``` Python library to plot a histogram of which months the scientists have birthdays in! Because it would take a long time for you to input the months of various scientists, you can use my scientist birthday JSON file. Just parse out the months (if you don’t know how, I suggest looking at the previous exercise or its solution) and draw your histogram.

If you are using a purely web-based interface for coding, this exercise won’t work for you, since it requires installing the bokeh Python package. Now might be a good time to install Python on your own computer.

## Solution:

````python
import json
from collections import Counter

# We need to first install bokeh since it doesn't come with Python. 
# In Linux, we do this by typing "pip3 install bokeh" in the terminal

f = open("bdays.json", "r")
bdays = json.load(f)
f.close()

months_as_string = {
    1: "January", 2: "February", 3: "March", 4: "April", 5: "May", 6: "June", 7: "July", 8: "August", 9: "September", 10: "October",
    11: "November", 12: "December"  
}

months = []

for b in bdays.values():
    month = int(b.split("/")[0])    
    months.append(months_as_string[month])

# Bokeh is a data visualizations library used for plotting
from bokeh.plotting import figure, show, output_file

# We specify an HTML file where the output will go
output_file("bdaysplot.html")

x_months = list(months_as_string.values())     # every month of the year (x-axis range)

# When plotting data we need to seperate our x variable and y variable into two seperate lists.
x = list(Counter(months).keys())               # list of only the months scientist have birthdays in
y = list(Counter(months).values())             # tally of months the scientists have bdays in 

# create a figure
p = figure(x_range = x_months)    # range of the x-axis needs to be specified

# create a histogram
p.vbar(x=x, top=y, width=0.8)

# add a plot title
p.title.text = "Frequency of Birthdays of Some Famous Scientists"
p.title.align = "center"

# angle the axis labels so they are easier to read
p.xaxis.major_label_orientation = 3.14/4  #pi/4

# render/show the plot
show(p)
````
